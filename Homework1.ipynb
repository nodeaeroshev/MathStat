Задание 1

```python
cards = 52
```

а)

```python
prob_tref = (13 / cards) * (12 / cards) * (11 / cards) * (10 / cards)

print(f"Вероятность все крести - {prob_tref:2f}")
```

б)

```python
prob_tuz = (4 / cards) + (4 / (cards - 1)) + (4 / (cards - 2)) + (4 / (cards - 3))
print(f"Вероятность хотя бы один туз - {prob_tuz:2f}")
```

Задание 3

```python
details = 15
painted = 9

prob_painted = (painted / details) * ((painted - 1) / (details - 1)) * ((painted - 2) / (details - 2))
print(f"Вероятность всех окрашенных деталей - {prob_painted:2f}")
```

Задание 4

```python
tickets = 100
winner = 2

prob_winner = (winner / tickets) * ((winner - 1) / (tickets - 1))
print(f"Вероятность выигрышных билетов - {prob_winner:2f}")
```
